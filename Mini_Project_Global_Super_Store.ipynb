Revenue Trends Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
revenue_trends = df_orders.groupby("Order Date")["Sales"].sum().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(revenue_trends["Order Date"], revenue_trends["Sales"], color="blue", alpha=0.6)
plt.xlabel("Year")
plt.ylabel("Total Sales ($)")
plt.title("Revenue Trends Over Time")
plt.grid(True)
plt.show()
```

Seasonal Sales Trends

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_orders["Order Date"] = pd.to_datetime(df_orders["Order Date"])
df_orders["Year"] = df_orders["Order Date"].dt.year
df_orders["Month"] = df_orders["Order Date"].dt.month
monthly_sales = df_orders.pivot_table(values="Sales", index="Year", columns="Month", aggfunc="sum", fill_value=0)
plt.figure(figsize=(12, 6))
sns.heatmap(monthly_sales, cmap="Blues", annot=True, fmt=".0f", linewidths=0.5)
plt.xlabel("Month")
plt.ylabel("Year")
plt.title("Seasonal Sales Trends (Heatmap)")
plt.show()
```

Top-Selling Products

```python
import seaborn as sns
import matplotlib.pyplot as plt

top_products = df.groupby('Product Name').agg({'Quantity': 'sum', 'Sales': 'sum'}).reset_index()
top_products = top_products.sort_values('Sales', ascending=False).head(10)

fig, ax = plt.subplots(1, 2, figsize=(14, 6))
sns.barplot(x='Quantity', y='Product Name', data=top_products, ax=ax[0], palette='Blues')
ax[0].set_title('Top 10 Products by Quantity Sold')
ax[0].set_xlabel('Total Quantity Sold')
ax[0].set_ylabel('Product Name')

sns.barplot(x='Sales', y='Product Name', data=top_products, ax=ax[1], palette='Greens')
ax[1].set_title('Top 10 Products by Sales Revenue')
ax[1].set_xlabel('Total Sales Revenue')
ax[1].set_ylabel('Product Name')

plt.tight_layout()
plt.show()
```

Revenue Trends Over Time

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Order Month'] = df['Order Date'].dt.to_period('M').astype(str)
df['Order Month'] = pd.to_datetime(df['Order Month'], format='%Y-%m')

monthly_trends = df.groupby('Order Month').agg({'Sales': 'sum'}).reset_index()

monthly_trends = monthly_trends.sort_values('Order Month')

plt.figure(figsize=(12, 6))
sns.lineplot(x='Order Month', y='Sales', data=monthly_trends, marker='o', color='purple')

plt.title('Monthly Revenue Trend', fontsize=18)
plt.xlabel('Order Month', fontsize=14)
plt.ylabel('Total Revenue (Sales)', fontsize=14)
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

Seasonal Sales Patterns

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df['Order Date'] = pd.to_datetime(df['Order Date'])

df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month

seasonal_trends = df.groupby(['Order Year', 'Order Month']).agg({'Sales': 'sum'}).reset_index()

seasonal_trends['Year-Month'] = pd.to_datetime(
    seasonal_trends['Order Year'].astype(str) + '-' + seasonal_trends['Order Month'].astype(str),
    format='%Y-%m'
)

seasonal_trends = seasonal_trends.sort_values('Year-Month')
plt.figure(figsize=(14, 6))
sns.lineplot(x='Year-Month', y='Sales', data=seasonal_trends, marker='o', color='royalblue')

plt.title('Seasonal Revenue Trends Over Time', fontsize=18)
plt.xlabel('Year-Month', fontsize=14)
plt.ylabel('Total Revenue (Sales)', fontsize=14)
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

Regional Sales Performance

```python
import pandas as pd
import plotly.express as px

region_sales = df.groupby('Region').agg({'Sales': 'sum'}).reset_index()

fig = px.bar(
    region_sales,
    x='Region',
    y='Sales',
    title='Total Revenue by Region',
    color='Sales',
    color_continuous_scale='blues',
    labels={'Sales': 'Total Revenue'}
)

fig.show()
```
